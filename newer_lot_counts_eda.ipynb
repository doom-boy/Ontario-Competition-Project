```python
import matplotlib.pyplot as plt
import pandas as pd

# --- Identify occupancy and capacity columns ---
occ_cols = [col for col in df.columns if "Occ" in col and "%" not in col]
cap_cols = [col for col in df.columns if ("Cap" in col or "Capacity" in col) and "%" not in col]

# --- Filter out Premium and Value lots ---
occ_cols = [c for c in occ_cols if not any(x in c for x in ["Prem", "Val", "Premium", "Value"])]
cap_cols = [c for c in cap_cols if not any(x in c for x in ["Prem", "Val", "Premium", "Value"])]

# --- Compute averages ---
avg_occ = df[occ_cols].mean()
avg_cap = df[cap_cols].mean()

# --- Pair occupancy and capacity columns ---
pairs = []
for occ in occ_cols:
    base_name = occ.replace("Occ", "").strip()
    cap_match = next((c for c in cap_cols if base_name in c), None)
    if cap_match:
        pairs.append((base_name.strip(), occ, cap_match))

# --- Build DataFrame ---
data = pd.DataFrame({
    "Lot": [name for name, _, _ in pairs],
    "Average Occupancy": [avg_occ[occ] for _, occ, _ in pairs],
    "Average Capacity": [avg_cap[cap] for _, _, cap in pairs]
})
data.set_index("Lot", inplace=True)

# --- Clean up lot names ---
data.index = data.index.str.replace("Gen", "", case=False, regex=False)
data.index = data.index.str.replace("General", "", case=False, regex=False)
data.index = data.index.str.strip()

# --- Add Total Facility ---
if "Total Facility" in df.columns:
    total_occ = df["Total Facility Occ"].mean()
    total_cap = df["Total Facility Cap"].mean()
    total_pct = (total_occ / total_cap) * 100
    data.loc["Total Facility"] = [total_occ, total_cap, total_pct]

# --- Convert to percentage of capacity ---
data["Average Occupancy %"] = (data["Average Occupancy"] / data["Average Capacity"]) * 100

# --- Define colors ---
colors = []
for lot in data.index:
    lot_lower = lot.lower()
    if any(num in lot_lower for num in ["2", "3", "4"]):
        colors.append("#5ca7fe")  # blue
    elif any(num in lot_lower for num in ["5", "6"]):
        colors.append("#dc288b")  # pink
    elif "total" in lot_lower:
        colors.append("lightgray")  # neutral
    else:
        colors.append("skyblue")  # fallback

# --- Plot ---
plt.figure(figsize=(10, 6))
x = range(len(data))
bars = plt.bar(x, data["Average Occupancy %"], color=colors, edgecolor="black", width=0.5)

# --- Add labels ---
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        f"{height:.1f}%",
        ha='center', va='bottom',
        fontsize=10,
        fontweight='bold'
    )

plt.title("Utilization Percentage by Lot (General Only)", fontsize=14, fontweight="bold")
plt.ylabel("Percentage of Capacity (%)")
plt.xticks(x, data.index, rotation=45, ha="right")
plt.ylim(0, 75)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Build the DataFrame as before ---
# (Assume 'data' already exists with index=Lot and 'Average Occupancy %' column)

# --- Reset index for seaborn ---
plot_data = data.reset_index()

# --- Define custom colors by lot ---
def lot_color(lot_name):
    lot_lower = lot_name.lower()
    if any(num in lot_lower for num in ["2", "3", "4"]):
        return "#5ca7fe"  # blue
    elif any(num in lot_lower for num in ["5", "6"]):
        return "#dc288b"  # pink
    elif "total" in lot_lower:
        return "#b0b0b0"  # neutral gray
    else:
        return "#8ecae6"  # fallback light blue

plot_data["color"] = plot_data["Lot"].apply(lot_color)

# --- Set seaborn style ---
sns.set(style="whitegrid")

# --- Plot ---
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    x="Lot", 
    y="Average Occupancy %", 
    data=plot_data, 
    palette=plot_data["color"].tolist(),
    edgecolor="black"
)

# --- Add value labels on top ---
for index, row in plot_data.iterrows():
    barplot.text(
        index, 
        row["Average Occupancy %"] + 0.5, 
        f"{row['Average Occupancy %']:.1f}%", 
        ha='center', 
        va='bottom', 
        fontweight='bold'
    )

plt.title("Utilization Percentage by Lot (General Only)", fontsize=16, fontweight="bold")
plt.ylabel("Percentage of Capacity (%)")
plt.ylim(0, 75)
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# --- Prepare data ---
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

col = 'Total Facility%'
df_rates = df[['Date', col]].melt(id_vars='Date', var_name='Rate Type', value_name='Rate')
df_rates = df_rates[df_rates['Rate'].notna()]

# Day of week
df_rates['Day of Week'] = df_rates['Date'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Compute average rate by day
data = df_rates.groupby('Day of Week')['Rate'].mean().reindex(day_order)

# Convert fraction to percentage
data_pct = data * 100

# --- Create subtle gradient colors ---
# Light to dark blue hex codes
light_blue = mpl.colors.to_rgb('#b0cbce')
dark_blue = mpl.colors.to_rgb('#4e909e')

# Normalize values for gradient
norm = mpl.colors.Normalize(vmin=data_pct.min(), vmax=data_pct.max())

# Linear interpolation for each bar color
colors = [(
    light_blue[0] + (dark_blue[0]-light_blue[0])*norm(val),
    light_blue[1] + (dark_blue[1]-light_blue[1])*norm(val),
    light_blue[2] + (dark_blue[2]-light_blue[2])*norm(val)
) for val in data_pct.values]

# --- Plot ---
plt.figure(figsize=(12,8))
bars = plt.bar(data_pct.index, data_pct.values, color=colors, edgecolor='black', width=0.7)

# Add percentage labels just above bars
for bar, val in zip(bars, data_pct.values):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        val + max(data_pct.values)*0.01,
        f'{val:.1f}%',
        ha='center',
        va='bottom',
        fontsize=12,
        weight='bold'
    )

# Styling
plt.title("Average Total Facility % by Day of Week (May to September 2025)", fontsize=16, weight='bold')
plt.ylabel("Percentage of Capacity (%)", fontsize=14)
plt.xlabel("Day of Week", fontsize=12)
sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.ylim(50, max(data_pct.values)*1.05)

plt.savefig(
    "daysofweek.png",  # filename
    dpi=300,                 # high resolution
    bbox_inches='tight'      # remove extra whitespace
)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

def plot_lot_bar(ax, df, col_name, title):
    """
    Plot average percentage by day of week for a given column on a given Axes.
    Uses subtle blue gradient from #b0cbce → #4e909e
    """
    df_rates = df[['Date', col_name]].melt(id_vars='Date', var_name='Rate Type', value_name='Rate')
    df_rates = df_rates[df_rates['Rate'].notna()]

    df_rates['Day of Week'] = df_rates['Date'].dt.day_name()
    day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
    data = df_rates.groupby('Day of Week')['Rate'].mean().reindex(day_order)
    data_pct = data * 100  # convert fraction to %

    # Gradient colors
    light_blue = mpl.colors.to_rgb('#b0cbce')
    dark_blue = mpl.colors.to_rgb('#4e909e')
    norm = mpl.colors.Normalize(vmin=data_pct.min(), vmax=data_pct.max())
    colors = [(
        light_blue[0] + (dark_blue[0]-light_blue[0])*norm(val),
        light_blue[1] + (dark_blue[1]-light_blue[1])*norm(val),
        light_blue[2] + (dark_blue[2]-light_blue[2])*norm(val)
    ) for val in data_pct.values]

    # Plot bars
    bars = ax.bar(data_pct.index, data_pct.values, color=colors, edgecolor='black', width=0.7)

    # Add percentage labels
    for bar, val in zip(bars, data_pct.values):
        ax.text(
            bar.get_x() + bar.get_width()/2,
            val + max(data_pct.values)*0.01,
            f'{val:.1f}%',
            ha='center',
            va='bottom',
            fontsize=11,
            weight='bold'
        )

    # Styling
    ax.set_title(title, fontsize=14, weight='bold')
    ax.set_ylabel("Percentage of Capacity (%)", fontsize=12)
    ax.set_xlabel("Day of Week", fontsize=12)
    ax.set_ylim(0, max(data_pct.values)*1.1)
    ax.grid(axis='y', linestyle='--', alpha=0.5)
    ax.set_xticklabels(data_pct.index, rotation=45)
    sns.despine(ax=ax)

# --- Create figure with 2 subplots ---
fig, axes = plt.subplots(2, 1, figsize=(12,14), constrained_layout=True)

plot_lot_bar(axes[0], df, 'Lot 2 Prem%', "Average Lot 2 Premium % by Day of Week")
plot_lot_bar(axes[1], df, 'Lot 4 Gen%', "Average Lot 4 General % by Day of Week")

# --- Save figure ---
plt.savefig("outliersweekdaydistribution.png", dpi=300, bbox_inches='tight')
plt.show()
```
