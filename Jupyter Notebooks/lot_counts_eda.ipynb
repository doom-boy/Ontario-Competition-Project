```python
#all lots
percent_cols = [
    'LOT 2 General % Capacity',
    'LOT 2 Premium % Capacity',
    'LOT 3 % Capacity',
    'LOT 4 General % Capacity',
    'LOT 4 Premium % Capacity',
    'LOT 5 % Capacity',
    'LOT 6 % Capacity'
]

# Keep only valid rows and convert date
df_clean = df.copy()
df_clean['Date'] = pd.to_datetime(df_clean['Date'])

# Melt into long format
melted = df_clean.melt(id_vars='Date', value_vars=percent_cols,
                       var_name='Lot', value_name='% Capacity')

# Remove invalid (-1) values and convert to percentage
melted = melted[melted['% Capacity'] != -1].copy()
melted['% Capacity'] *= 100

# Group by week and lot, averaging each week's values
weekly = (
    melted
    .groupby([pd.Grouper(key='Date', freq='W'), 'Lot'])['% Capacity']
    .mean()
    .reset_index()
)

# Create FacetGrid (each lot gets its own subplot)
g = sns.FacetGrid(weekly, col='Lot', col_wrap=3, height=3.5, sharey=False)
g.map_dataframe(sns.lineplot, x='Date', y='% Capacity', color='seagreen', marker='o')

# Adjust labels and layout
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Week", "% Capacity")
for ax in g.axes.flat:
    ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Select relevant revenue columns
revenue_cols = ['Total CC Revenue', 'Total Cash Revenue', 'Total Revenue']

# Prepare dataframe
df_rev = df.copy()
df_rev['Date'] = pd.to_datetime(df_rev['Date'])

# Melt into long format for FacetGrid
melted_rev = df_rev.melt(id_vars='Date', value_vars=revenue_cols,
                         var_name='Revenue Type', value_name='Revenue')

# Remove invalid or negative values
melted_rev = melted_rev[melted_rev['Revenue'] >= 0].copy()

# Group by week (sum makes more sense for revenue than mean)
weekly_rev = (
    melted_rev
    .groupby([pd.Grouper(key='Date', freq='W'), 'Revenue Type'])['Revenue']
    .sum()
    .reset_index()
)

# Create FacetGrid for weekly revenue
g = sns.FacetGrid(weekly_rev, col='Revenue Type', col_wrap=3, height=3.5, sharey=False)
g.map_dataframe(sns.lineplot, x='Date', y='Revenue', color='teal', marker='o')

# Adjust labels and style
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Week", "Revenue ($)")
for ax in g.axes.flat:
    ax.grid(True, linestyle='--', alpha=0.4)
    ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Define your columns
occupied_cols = [
    'LOT 2 General Occupied',
    'LOT 2 Premium Occupied',
    'LOT 3 Occupied',
    'LOT 4 General Occupied',
    'LOT 4 Premium Occupied',
    'LOT 5 Occupied',
    'LOT 6 Occupied'
]

capacity_cols = [
    'LOT 2 General Spaces',
    'LOT 2 Premium Spaces',
    'LOT 3 Spaces',
    'LOT 4 General Spaces',
    'LOT 4 Premium Spaces',
    'LOT 5 Spaces',
    'LOT 6 Spaces'
]

# Compute totals conditionally
def compute_totals(row):
    if row['LOT 6 Occupied'] == -1:
        occ_total = row[[c for c in occupied_cols if 'LOT 6' not in c]].sum()
        cap_total = row[[c for c in capacity_cols if 'LOT 6' not in c]].sum()
    else:
        occ_total = row[occupied_cols].sum()
        cap_total = row[capacity_cols].sum()
    return pd.Series({'Total Occupied': occ_total, 'Total Spaces': cap_total})

# Apply to each row
totals = df.apply(compute_totals, axis=1)

# Add to DataFrame
df['Total Occupied'] = totals['Total Occupied']
df['Total Spaces'] = totals['Total Spaces']
df['Utilization Rate'] = (df['Total Occupied'] / df['Total Spaces']) * 100

# Optional: round utilization for readability
df['Utilization Rate'] = df['Utilization Rate'].round(2)

# Preview result
df[['Total Occupied', 'Total Spaces', 'Utilization Rate']].head()
```

```python
# --- 1️⃣ Prepare data ---
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# Filter out invalid utilization rows (in case Lot 6 was -1)
util_df = df[df['Utilization Rate'].notna()].copy()

# --- 2️⃣ DAILY VIEW ---
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=util_df,
    x='Date',
    y='Utilization Rate',
    color='royalblue',
    marker='o'
)
plt.title('Daily Total Utilization Rate')
plt.xlabel('Date')
plt.ylabel('Utilization Rate (%)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# --- 3️⃣ WEEKLY VIEW ---
weekly = (
    util_df
    .groupby(pd.Grouper(key='Date', freq='W'))['Utilization Rate']
    .mean()
    .reset_index()
)

plt.figure(figsize=(10, 5))
sns.lineplot(
    data=weekly,
    x='Date',
    y='Utilization Rate',
    color='seagreen',
    marker='o'
)
plt.title('Weekly Average Utilization Rate')
plt.xlabel('Week')
plt.ylabel('Average Utilization Rate (%)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Prepare data ---
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter for March 2022 → March 2023
start_date = '2022-03-01'
end_date = '2023-03-31'
df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)].copy()

# Add Month column (abbreviated)
df_filtered['Month'] = df_filtered['Date'].dt.strftime('%b')

# Select all % Capacity columns
pct_cols = [col for col in df_filtered.columns if '% Capacity' in col]

# Compute monthly averages
df_monthly = df_filtered.groupby('Month')[pct_cols].mean().reindex(
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
)

# --- Plot using FacetGrid style ---
n_cols = 3  # number of subplots per row
n_rows = -(-len(pct_cols) // n_cols)  # ceiling division
fig, axes = plt.subplots(n_rows, n_cols, figsize=(6*n_cols, 4*n_rows), constrained_layout=True)
axes = axes.flatten()

colors = sns.color_palette("Set2", len(pct_cols))

for i, col in enumerate(pct_cols):
    ax = axes[i]
    ax.bar(df_monthly.index, df_monthly[col], color=colors[i % len(colors)], edgecolor='black')
    ax.set_title(col, fontsize=12, weight='bold')
    ax.set_ylim(0, 1)
    ax.set_xlabel('Month')
    ax.set_ylabel('Average % Capacity')
    sns.despine(ax=ax)
    ax.grid(False)

# Remove any empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Prepare data ---
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter for March 2022 → March 2023
start_date = '2022-03-01'
end_date = '2023-03-31'
df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)].copy()

# Select all % Capacity columns
pct_cols = [col for col in df_filtered.columns if '% Capacity' in col]

# Compute average occupancy per lot
avg_lot_occupancy = df_filtered[pct_cols].mean().sort_values(ascending=False)

# --- Plot ---
plt.figure(figsize=(10, 5))
sns.barplot(
    x=avg_lot_occupancy.index,
    y=avg_lot_occupancy.values,
    palette='Set2',
    edgecolor='black'
)

plt.title('Average Occupancy % per Lot (Mar 2022 – Mar 2023)', fontsize=14, weight='bold')
plt.ylabel('Average % Capacity')
plt.xlabel('Lot')
plt.ylim(0, 1)
plt.xticks(rotation=45, ha='right')
sns.despine()
plt.grid(False)

# --- Add labels above bars ---
for i, val in enumerate(avg_lot_occupancy.values):
    plt.text(i, val + 0.02, f"{val*100:.1f}%", ha='center', va='bottom', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()
```
