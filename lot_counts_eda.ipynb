```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select % Capacity columns
percent_cols = [
    'LOT 2 General % Capacity',
    'LOT 2 Premium % Capacity',
    'LOT 3 % Capacity',
    'LOT 4 General % Capacity',
    'LOT 4 Premium % Capacity',
    'LOT 5 % Capacity',
    'LOT 6 % Capacity'
]

# Melt into long format for FacetGrid
melted_pct = df[percent_cols].melt(var_name='Lot', value_name='% Capacity')

# Convert from decimal (e.g. 0.7) to percentage (e.g. 70)
melted_pct['% Capacity'] = melted_pct['% Capacity'] * 100

# Create facet grid
g = sns.FacetGrid(melted_pct, col='Lot', col_wrap=3, sharex=False, sharey=False)
g.map_dataframe(sns.histplot, x='% Capacity', bins=20, color='skyblue', edgecolor='black')

# Adjust labels
g.set_titles(col_template="{col_name}")
g.set_axis_labels("% Capacity", "Count")
plt.tight_layout()
plt.show()
```

```python
# Prepare data
lot2 = df[df['LOT 2 General % Capacity'] != -1].copy()
lot2['Date'] = pd.to_datetime(lot2['Date'])
lot2 = lot2.sort_values('Date')

# Convert to percentage
lot2['LOT 2 General % Capacity'] *= 100

# Group by week (using Sunday as end of week)
weekly = (
    lot2
    .groupby(pd.Grouper(key='Date', freq='W'))['LOT 2 General % Capacity']
    .mean()
    .reset_index()
)

# Plot weekly averages
plt.figure(figsize=(10, 5))
sns.lineplot(data=weekly, x='Date', y='LOT 2 General % Capacity', color='seagreen', marker='o')
plt.title('Lot 2 General % Capacity (Weekly Average)')
plt.xlabel('Week')
plt.ylabel('% Capacity')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#all lots
percent_cols = [
    'LOT 2 General % Capacity',
    'LOT 2 Premium % Capacity',
    'LOT 3 % Capacity',
    'LOT 4 General % Capacity',
    'LOT 4 Premium % Capacity',
    'LOT 5 % Capacity',
    'LOT 6 % Capacity'
]

# Keep only valid rows and convert date
df_clean = df.copy()
df_clean['Date'] = pd.to_datetime(df_clean['Date'])

# Melt into long format
melted = df_clean.melt(id_vars='Date', value_vars=percent_cols,
                       var_name='Lot', value_name='% Capacity')

# Remove invalid (-1) values and convert to percentage
melted = melted[melted['% Capacity'] != -1].copy()
melted['% Capacity'] *= 100

# Group by week and lot, averaging each week's values
weekly = (
    melted
    .groupby([pd.Grouper(key='Date', freq='W'), 'Lot'])['% Capacity']
    .mean()
    .reset_index()
)

# Create FacetGrid (each lot gets its own subplot)
g = sns.FacetGrid(weekly, col='Lot', col_wrap=3, height=3.5, sharey=False)
g.map_dataframe(sns.lineplot, x='Date', y='% Capacity', color='seagreen', marker='o')

# Adjust labels and layout
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Week", "% Capacity")
for ax in g.axes.flat:
    ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Select relevant revenue columns
revenue_cols = ['Total CC Revenue', 'Total Cash Revenue', 'Total Revenue']

# Prepare dataframe
df_rev = df.copy()
df_rev['Date'] = pd.to_datetime(df_rev['Date'])

# Melt into long format for FacetGrid
melted_rev = df_rev.melt(id_vars='Date', value_vars=revenue_cols,
                         var_name='Revenue Type', value_name='Revenue')

# Remove invalid or negative values
melted_rev = melted_rev[melted_rev['Revenue'] >= 0].copy()

# Group by week (sum makes more sense for revenue than mean)
weekly_rev = (
    melted_rev
    .groupby([pd.Grouper(key='Date', freq='W'), 'Revenue Type'])['Revenue']
    .sum()
    .reset_index()
)

# Create FacetGrid for weekly revenue
g = sns.FacetGrid(weekly_rev, col='Revenue Type', col_wrap=3, height=3.5, sharey=False)
g.map_dataframe(sns.lineplot, x='Date', y='Revenue', color='teal', marker='o')

# Adjust labels and style
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Week", "Revenue ($)")
for ax in g.axes.flat:
    ax.grid(True, linestyle='--', alpha=0.4)
    ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Prepare data
df_pct = df.copy()
df_pct['Date'] = pd.to_datetime(df_pct['Date'])

# Select relevant columns
percent_cols = [
    'LOT 2 General % Capacity',
    'LOT 2 Premium % Capacity',
    'LOT 3 % Capacity',
    'LOT 4 General % Capacity',
    'LOT 4 Premium % Capacity',
    'LOT 5 % Capacity',
    'LOT 6 % Capacity'
]

# Melt
melted = df_pct.melt(id_vars='Date', value_vars=percent_cols,
                     var_name='Lot', value_name='% Capacity')

# Clean and convert to %
melted = melted[melted['% Capacity'] != -1].copy()
melted['% Capacity'] *= 100

# Extract week label (like '2023-03-27' for week start)
melted['Week'] = melted['Date'].dt.to_period('W').apply(lambda r: r.start_time)

# Create FacetGrid: one plot per lot, colored by week
g = sns.FacetGrid(melted, col='Lot', col_wrap=3, sharex=False, sharey=False, height=3.5)
g.map_dataframe(sns.histplot, x='% Capacity', hue='Week', bins=20, multiple='dodge', edgecolor='black')

# Adjust legend and labels
g.set_titles(col_template="{col_name}")
g.set_axis_labels("% Capacity", "Count")
g.add_legend(title="Week")
plt.tight_layout()
plt.show()
```
