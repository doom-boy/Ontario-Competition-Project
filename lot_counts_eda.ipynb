```python
plt.hist(df['LOT 2 General Occupied'], color= "skyblue", edgecolor = "black", bins = 50)
plt.title("Lot 2 Occupancies Distribution")
plt.xlabel("Occupancies")
plt.ylabel("Count")
```

```python
# Select the relevant columns
cols = [
    'LOT 2 General Occupied',
    'LOT 2 Premium Occupied',
    'LOT 3 Occupied',
    'LOT 4 General Occupied',
    'LOT 4 Premium Occupied',
    'LOT 5 Occupied',
    'LOT 6 Occupied'
]

# Melt into long format for FacetGrid
melted = df[cols].melt(var_name='Lot', value_name='Occupancies')

# Create facet grid
g = sns.FacetGrid(melted, col='Lot', col_wrap=3, sharex=False, sharey=False)
g.map_dataframe(sns.histplot, x='Occupancies', bins=20, color='skyblue', edgecolor='black')

# Adjust labels
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Occupancies", "Count")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select % Capacity columns
percent_cols = [
    'LOT 2 General % Capacity',
    'LOT 2 Premium % Capacity',
    'LOT 3 % Capacity',
    'LOT 4 General % Capacity',
    'LOT 4 Premium % Capacity',
    'LOT 5 % Capacity',
    'LOT 6 % Capacity'
]

# Melt into long format for FacetGrid
melted_pct = df[percent_cols].melt(var_name='Lot', value_name='% Capacity')

# Convert from decimal (e.g. 0.7) to percentage (e.g. 70)
melted_pct['% Capacity'] = melted_pct['% Capacity'] * 100

# Create facet grid
g = sns.FacetGrid(melted_pct, col='Lot', col_wrap=3, sharex=False, sharey=False)
g.map_dataframe(sns.histplot, x='% Capacity', bins=20, color='skyblue', edgecolor='black')

# Adjust labels
g.set_titles(col_template="{col_name}")
g.set_axis_labels("% Capacity", "Count")
plt.tight_layout()
plt.show()
```
